```python
'''–ü—Ä–æ—Å—Ç–æ –∑–¥—Ä–∞–≤—Å—Ç–≤—É–π, –ø—Ä–æ—Å—Ç–æ –∫–∞–∫ –¥–µ–ª–∞
–£–º–µ–Ω–∏–µ –≤–µ—Å—Ç–∏ –¥–∏–∞–ª–æ–≥ ‚Äî –≤–∞–∂–Ω—ã–π –Ω–∞–≤—ã–∫ –¥–ª—è –≤–æ—Å–ø–∏—Ç–∞–Ω–Ω–æ–≥–æ —á–µ–ª–æ–≤–µ–∫–∞.

–ù–∞–ø–∏—à–∏—Ç–µ –¥–∏–∞–ª–æ–≥–æ–≤—É—é –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–Ω–∞—á–∞–ª–∞ –ø–æ–∑–Ω–∞–∫–æ–º–∏—Ç—Å—è —Å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º, –∞ –∑–∞—Ç–µ–º –ø–æ–∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç—Å—è –µ–≥–æ –Ω–∞—Å—Ç—Ä–æ–µ–Ω–∏–µ–º.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ –∏–º—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.
–í–æ –≤—Ç–æ—Ä–æ–π ‚Äî –æ—Ç–≤–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å: ¬´—Ö–æ—Ä–æ—à–æ¬ª –∏–ª–∏ ¬´–ø–ª–æ—Ö–æ¬ª.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –≤–æ–ø—Ä–æ—Å ¬´–ö–∞–∫ –í–∞—Å –∑–æ–≤—É—Ç?¬ª
–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ ‚Äî ¬´–ó–¥—Ä–∞–≤—Å—Ç–≤—É–π—Ç–µ, %username%!¬ª
–í —Ç—Ä–µ—Ç—å–µ–π —Å—Ç—Ä–æ–∫–µ ‚Äî –≤–æ–ø—Ä–æ—Å ¬´–ö–∞–∫ –¥–µ–ª–∞?¬ª
–í —á–µ—Ç–≤—ë—Ä—Ç–æ–π —Å—Ç—Ä–æ–∫–µ —Ä–µ–∞–∫—Ü–∏—è –Ω–∞ –æ—Ç–≤–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è:

–µ—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –æ—Ç–≤–µ—Ç–∏–ª ¬´—Ö–æ—Ä–æ—à–æ¬ª, —Å–ª–µ–¥—É–µ—Ç –≤—ã–≤–µ—Å—Ç–∏ —Å–æ–æ–±—â–µ–Ω–∏–µ ¬´–Ø –∑–∞ –≤–∞—Å —Ä–∞–¥–∞!¬ª;'''


print('–ö–∞–∫ –í–∞—Å –∑–æ–≤—É—Ç?')
name = input()
print(f'–ó–¥—Ä–∞–≤—Å—Ç–≤—É–π—Ç–µ, {name}!\n–ö–∞–∫ –¥–µ–ª–∞?')
status = input()
if '—Ö–æ—Ä–æ—à–æ' in status:
    print('–Ø –∑–∞ –≤–∞—Å —Ä–∞–¥–∞!')
elif '–ø–ª–æ—Ö–æ' in status:
    print('–í—Å—ë –Ω–∞–ª–∞–¥–∏—Ç—Å—è!')
```

```python
'''–ö—Ç–æ –±—ã—Å—Ç—Ä–µ–µ?
–í –≥–ª–∞–≤–Ω–æ–π –≤–µ–ª–æ–≥–æ–Ω–∫–µ –≥–æ–¥–∞ —É—á–∞—Å—Ç–≤—É–µ—Ç –±–æ–ª–µ–µ —Ç—ã—Å—è—á–∏ –≥–æ–Ω—â–∏–∫–æ–≤. –ò–º –ø—Ä–µ–¥—Å—Ç–æ–∏—Ç –ø—Ä–æ–π—Ç–∏ —Ç—Ä–∞—Å—Å—É –¥–ª–∏–Ω–æ–π 43872–º. 
–°–∞–º–∞—è —Å–ª–æ–∂–Ω–∞—è –∏ –æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–∞—è –∑–∞–¥–∞—á–∞ ‚Äî –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–±–µ–¥–∏—Ç–µ–ª—è.

–ù–∞–º –∏–∑–≤–µ—Å—Ç–Ω—ã —Å—Ä–µ–¥–Ω–∏–µ —Å–∫–æ—Ä–æ—Å—Ç–∏ –¥–≤—É—Ö —Ñ–∞–≤–æ—Ä–∏—Ç–æ–≤ ‚Äî –ü–µ—Ç–∏ –∏ –í–∞—Å–∏. –ü–æ–º–æ–≥–∏—Ç–µ –≤—ã—è—Å–Ω–∏—Ç—å, –∫—Ç–æ –∏–∑ –Ω–∏—Ö –ø—Ä–∏—à—ë–ª –∫ —Ñ–∏–Ω–∏—à—É –ø–µ—Ä–≤—ã–º.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–∞ —Å—Ä–µ–¥–Ω—è—è —Å–∫–æ—Ä–æ—Å—Ç—å –ü–µ—Ç–∏.
–í–æ –≤—Ç–æ—Ä–æ–π ‚Äî –í–∞—Å–∏.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–ò–º—è –ø–æ–±–µ–¥–∏—Ç–µ–ª—è –≥–æ–Ω–∫–∏.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ
–ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ –ø–æ–±–µ–¥–∏—Ç–µ–ª–µ–º —Å—Ç–∞–ª —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω.'''


p_speed = int(input())
v_speed = int(input())
if p_speed > v_speed:
    print('–ü–µ—Ç—è')
else:
    print('–í–∞—Å—è')
```

```python
'''–ö—Ç–æ –±—ã—Å—Ç—Ä–µ–µ –Ω–∞ —ç—Ç–æ—Ç —Ä–∞–∑?
–í–Ω–æ–≤—å –≤–µ–ª–æ–≥–æ–Ω—â–∏–∫–∏ —Å–æ–±—Ä–∞–ª–∏—Å—å —É–∑–Ω–∞—Ç—å, –∫—Ç–æ –∏–∑ –Ω–∏—Ö –±—ã—Å—Ç—Ä–µ–µ. –ò–º –ø—Ä–µ–¥—Å—Ç–æ–∏—Ç –ø—Ä–æ–π—Ç–∏ —Ç—Ä–∞—Å—Å—É –¥–ª–∏–Ω–æ–π 43872–º, 
–∏ –Ω–∞–º –Ω—É–∂–Ω–æ –≤–Ω–æ–≤—å –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –ø–æ–±–µ–¥–∏—Ç–µ–ª—è.

–ù–∞ —ç—Ç–æ—Ç —Ä–∞–∑ –Ω–∞–º –∏–∑–≤–µ—Å—Ç–Ω—ã —Å—Ä–µ–¥–Ω–∏–µ —Å–∫–æ—Ä–æ—Å—Ç–∏ —Ç—Ä—ë—Ö —Ñ–∞–≤–æ—Ä–∏—Ç–æ–≤ ‚Äî –ü–µ—Ç–∏, –í–∞—Å–∏ –∏ –¢–æ–ª–∏. –ö—Ç–æ –∏–∑ –Ω–∏—Ö –ø—Ä–∏—à—ë–ª –∫ —Ñ–∏–Ω–∏—à—É –ø–µ—Ä–≤—ã–º?

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–∞ —Å—Ä–µ–¥–Ω—è—è —Å–∫–æ—Ä–æ—Å—Ç—å –ü–µ—Ç–∏.
–í–æ –≤—Ç–æ—Ä–æ–π ‚Äî –í–∞—Å–∏.
–í —Ç—Ä–µ—Ç—å–µ–π ‚Äî –¢–æ–ª–∏.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–ò–º—è –ø–æ–±–µ–¥–∏—Ç–µ–ª—è –≥–æ–Ω–∫–∏.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ
–ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ –ø–æ–±–µ–¥–∏—Ç–µ–ª–µ–º —Å—Ç–∞–ª —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω.'''


p_speed = int(input())
v_speed = int(input())
t_speed = int(input())
if v_speed < p_speed > t_speed:
    print('–ü–µ—Ç—è')
elif p_speed < v_speed > t_speed:
    print('–í–∞—Å—è')
else:
    print('–¢–æ–ª—è')
```

```python
'''–°–ø–∏—Å–æ–∫ –ø–æ–±–µ–¥–∏—Ç–µ–ª–µ–π
–î–ª–∏–Ω–∞ —Ç—Ä–∞—Å—Å—ã ‚Äî 43872–º, –∏ –∑—Ä–∏—Ç–µ–ª–∏ —Ö–æ—Ç—è—Ç —É–∑–Ω–∞—Ç—å –∏–º—è –ø–æ–±–µ–¥–∏—Ç–µ–ª—è.

–ù–∞–º –∏–∑–≤–µ—Å—Ç–Ω—ã —Å—Ä–µ–¥–Ω–∏–µ —Å–∫–æ—Ä–æ—Å—Ç–∏ —Ç—Ä—ë—Ö —Ñ–∞–≤–æ—Ä–∏—Ç–æ–≤ ‚Äì –ü–µ—Ç–∏, –í–∞—Å–∏ –∏ –¢–æ–ª–∏. –ü–æ–º–æ–≥–∏—Ç–µ –ø–æ–¥–≤–µ—Å—Ç–∏ –∏—Ç–æ–≥–∏ –≥–æ–Ω–∫–∏.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–∞ —Å—Ä–µ–¥–Ω—è—è —Å–∫–æ—Ä–æ—Å—Ç—å –ü–µ—Ç–∏.
–í–æ –≤—Ç–æ—Ä–æ–π ‚Äî –í–∞—Å–∏.
–í —Ç—Ä–µ—Ç—å–µ–π ‚Äî –¢–æ–ª–∏.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–ò–º–µ–Ω–∞ –ø–æ–±–µ–¥–∏—Ç–µ–ª–µ–π –≤ –ø–æ—Ä—è–¥–∫–µ –∑–∞–Ω—è—Ç—ã—Ö –º–µ—Å—Ç.'''


p = int(input())
v = int(input())
t = int(input())
mid = (p + v + t) - (max(p, v, t) + min(p, v, t))
if p == max(p, v, t):
    vin = '1. –ü–µ—Ç—è'
elif v == max(p, v, t):
    vin = '1. –í–∞—Å—è'
elif t == max(p, v, t):
    vin = '1. –¢–æ–ª—è'
if p == mid:
    sec = '2. –ü–µ—Ç—è'
elif v == mid:
    sec = '2. –í–∞—Å—è'
elif t == mid:
    sec = '2. –¢–æ–ª—è'
if p == min(p, v, t):
    los = '3. –ü–µ—Ç—è'
elif v == min(p, v, t):
    los = '3. –í–∞—Å—è'
elif t == min(p, v, t):
    los = '3. –¢–æ–ª—è'
print(vin)
print(sec)
print(los)
```

```python
'''–Ø–±–ª–æ–∫–∏
–£ –ü–µ—Ç–∏ –±—ã–ª–æ 7 —è–±–ª–æ–∫, –∞ —É –í–∞—Å–∏ 6.
–ó–∞—Ç–µ–º –ü–µ—Ç—è –æ—Ç–¥–∞–ª 3 —è–±–ª–æ–∫–∞ –í–∞—Å–µ, –∞ —É –¢–æ–ª–∏ –≤–∑—è–ª 2 —è–±–ª–æ–∫–∞.
–í–∞—Å—è –ø–æ–ø—Ä–æ—Å–∏–ª —É –¢–æ–ª–∏ 5 —è–±–ª–æ–∫, –Ω–æ –æ–Ω –æ—Ç–¥–∞–ª –ì–µ–Ω–µ 2.
–ó–∞—Ç–µ–º –î–∏–º–∞ –¥–∞–ª –ü–µ—Ç–µ N —è–±–ª–æ–∫, –∞ –í–∞—Å–µ M.
–¢–∞–∫ —É –∫–æ–≥–æ –≤ –∏—Ç–æ–≥–µ —è–±–ª–æ–∫ –±–æ–ª—å—à–µ ‚Äî —É –ü–µ—Ç–∏ –∏–ª–∏ –í–∞—Å–∏?

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ N.
–í–æ –≤—Ç–æ—Ä–æ–π ‚Äî M.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–ò–º—è —Ä–µ–±—ë–Ω–∫–∞, —É –∫–æ—Ç–æ—Ä–æ–≥–æ –±–æ–ª—å—à–µ —è–±–ª–æ–∫.'''


p = 7 - 3 + 2
v = 6 + 3 + 5 - 2
d_p = int(input())
d_v = int(input())
t_p = p + d_p
t_v = v + d_v
if t_p > t_v:
    print('–ü–µ—Ç—è')
else:
    print('–í–∞—Å—è')
```

```python
'''–°–∏–ª–∞ –ø—Ä–æ–∫—Ä–∞—Å—Ç–∏–Ω–∞—Ü–∏–∏
–í–∞—Å—è –ª—é–±–∏—Ç –ø–æ–ª–µ–Ω–∏—Ç—å—Å—è. –û—Å–æ–±–µ–Ω–Ω–æ –µ–º—É –Ω—Ä–∞–≤–∏—Ç—Å—è, –∫–æ–≥–¥–∞ –≤ –≥–æ–¥—É –ø–æ—è–≤–ª—è–µ—Ç—Å—è —Ç–∞–∫–æ–π –ª–∏—à–Ω–∏–π –¥–µ–Ω—ë–∫, 
–∫–æ—Ç–æ—Ä–æ–≥–æ –æ–±—ã—á–Ω–æ –Ω–µ –±—ã–≤–∞–µ—Ç. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–º–æ–∂–µ—Ç –í–∞—Å–µ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, 
—É–¥–∞—Å—Ç—Å—è –ª–∏ –µ–º—É –ø–æ–±–µ–∑–¥–µ–ª—å–Ω–∏—á–∞—Ç—å –≤ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–º –≥–æ–¥—É –∏–ª–∏ –Ω–µ—Ç.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–û–¥–Ω–æ —á–∏—Å–ª–æ ‚Äî –≥–æ–¥.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–û–¥–Ω–æ —Å–ª–æ–≤–æ ¬´YES¬ª (—É–¥–∞—Å—Ç—Å—è) –∏–ª–∏ ¬´NO¬ª (–Ω–µ —É–¥–∞—Å—Ç—Å—è).'''


year = int(input())
if year % 4 == 0 and year % 100 != 0:
    print('YES')
elif year % 100 == 0 and year % 400 == 0:
    print('YES')
else:
    print('NO')
```

```python
'''–ê —Ä–æ–∑–∞ —É–ø–∞–ª–∞ –Ω–∞ –ª–∞–ø—É –ê–∑–æ—Ä–∞
–°—É—â–µ—Å—Ç–≤—É—é—Ç —Ç–∞–∫–æ–µ –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ–µ –ø–æ–Ω—è—Ç–∏–µ –∫–∞–∫ –ø–∞–ª–∏–Ω–¥—Ä–æ–º ‚Äî —á–∏—Å–ª–æ, —Å–ª–æ–≤–æ, –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–µ –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ, 
–∫–æ—Ç–æ—Ä–æ–µ –∏ —Å–ª–µ–≤–∞-–Ω–∞–ø—Ä–∞–≤–æ, –∏ —Å–ø—Ä–∞–≤–∞-–Ω–∞–ª–µ–≤–æ —á–∏—Ç–∞–µ—Ç—Å—è –æ–¥–∏–Ω–∞–∫–æ–≤–æ.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ–≤–µ—Ä—è–µ—Ç, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —á–∏—Å–ª–æ –ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–û–¥–Ω–æ —á–µ—Ç—ã—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
YES –µ—Å–ª–∏ —á–∏—Å–ª–æ —è–≤–ª—è–µ—Ç—Å—è –ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º, –∏–Ω–∞—á–µ ‚Äî NO.'''


x = int(input())
a = x // 1000
b = x // 100 % 10
c = x // 10 % 10
d = x % 10
y = int(f'{d}{c}{b}{a}')
if x == y:
    print('YES')
else:
    print('NO')
```

```python
'''–ó–∞–π–∫–∞ ‚Äî 1
–í –¥–æ–ª–≥–æ–π –¥–æ—Ä–æ–≥–µ –¥–µ—Ç–∏ –º–æ–≥—É—Ç –∫–∞–ø—Ä–∏–∑–Ω–∏—á–∞—Ç—å, –ø–æ—ç—Ç–æ–º—É —Ä–æ–¥–∏—Ç–µ–ª–∏ –∏—Ö —Ä–∞–∑–≤–ª–µ–∫–∞—é—Ç —Å –ø–æ–º–æ—â—å—é –∏–≥—Ä. 
–û–¥–Ω–∞ –∏–∑ –Ω–∏—Ö ‚Äî –ø–æ–∏—Å–∫ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∑–≤–µ—Ä—É—à–µ–∫ –≤ –ø—Ä–∏–¥–æ—Ä–æ–∂–Ω–æ–π —Ä–∞—Å—Ç–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏.

–î–∞–≤–∞–π—Ç–µ –Ω–µ–º–Ω–æ–≥–æ –ø–æ–∏–≥—Ä–∞–µ–º –∏ –≤—ã—è—Å–Ω–∏–º, –Ω–µ —Å–ø—Ä—è—Ç–∞–ª—Å—è –ª–∏ –∑–∞–π–∫–∞ –≤–æ –≤–≤–µ–¥—ë–Ω–Ω–æ–º –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–∏.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–°—Ç—Ä–æ–∫–∞ –æ–ø–∏—Å—ã–≤–∞—é—â–∞—è –ø—Ä–∏–¥–æ—Ä–æ–∂–Ω—É—é –º–µ—Å—Ç–Ω–æ—Å—Ç—å.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
YES ‚Äî –µ—Å–ª–∏ –≤ —ç—Ç–æ–π –º–µ—Å—Ç–Ω–æ—Å—Ç–∏ –µ—Å—Ç—å –∑–∞–π–∫–∞, –∏–Ω–∞—á–µ ‚Äî NO.'''


wid = input()
if '–∑–∞–π–∫–∞' in wid:
    print('YES')
else:
    print('NO')
```

```python
'''–ü–µ—Ä–≤–æ–º—É –∏–≥—Ä–æ–∫—É –ø—Ä–∏–≥–æ—Ç–æ–≤–∏—Ç—å—Å—è
–í–æ –º–Ω–æ–≥–∏—Ö –∏–≥—Ä–∞—Ö –ø–æ—Ä—è–¥–æ–∫ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –±—Ä–æ—Å–∫–æ–º –∫—É–±–∏–∫–∞ –∏–ª–∏ –º–æ–Ω–µ—Ç–∫–∏, ‚Äî –∞ –≤ –Ω–∞—à–µ–π –ø–µ—Ä–≤—ã–º —Ö–æ–¥–∏—Ç —Ç–æ—Ç, 
—á—å—ë –∏–º—è –ª–µ–∫—Å–∏–∫–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –º–µ–Ω—å—à–µ.

–û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, –∫—Ç–æ –∏–∑ –∏–≥—Ä–æ–∫–æ–≤ –±—É–¥–µ—Ç —Ö–æ–¥–∏—Ç—å –ø–µ—Ä–≤—ã–º.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–¢—Ä–∏ –∏–º–µ–Ω–∏ –∏–≥—Ä–æ–∫–æ–≤, –∫–∞–∂–¥–æ–µ –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –∑–∞–ø–∏—Å–∞–Ω–æ —Å –Ω–æ–≤–æ–π —Å—Ç—Ä–æ–∫–∏.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–ò–º—è –∏–≥—Ä–æ–∫–∞, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç —Ö–æ–¥–∏—Ç—å –ø–µ—Ä–≤—ã–º.'''


name1 = input()
name2 = input()
name3 = input()
print(min(name1, name2, name3))
```

```python
'''–õ—É—á—à–∞—è –∑–∞—â–∏—Ç–∞ ‚Äî —à–∏—Ñ—Ä–æ–≤–∞–Ω–∏–µ
–ö–æ–ª—è –∏—Å–ø—É–≥–∞–ª—Å—è, —á—Ç–æ –ê–Ω—è –ø–æ–¥—Å–º–æ—Ç—Ä–∏—Ç –≤—Å–µ –µ–≥–æ –ø–∞—Ä–æ–ª–∏ –≤ –±–ª–æ–∫–Ω–æ—Ç–µ, –∏ —Ä–µ—à–∏–ª –∏—Ö –∑–∞—à–∏—Ñ—Ä–æ–≤–∞—Ç—å. 
–î–ª—è —ç—Ç–æ–≥–æ –æ–Ω –±–µ—Ä—ë—Ç –∏–∑–Ω–∞—á–∞–ª—å–Ω—ã–π –ø–∞—Ä–æ–ª—å ‚Äî —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ ‚Äî –∏ –ø–æ –Ω–µ–º—É —Å—Ç—Ä–æ–∏—Ç –Ω–æ–≤–æ–µ —á–∏—Å–ª–æ –ø–æ —Å–ª–µ–¥—É—é—â–∏–º –ø—Ä–∞–≤–∏–ª–∞–º:

–Ω–∞—Ö–æ–¥–∏—Ç—Å—è —Å—É–º–º–∞ —Ü–∏—Ñ—Ä, —Å—Ç–æ—è—â–∏—Ö –≤ –¥–≤—É—Ö –º–ª–∞–¥—à–∏—Ö —Ä–∞–∑—Ä—è–¥–∞—Ö (–¥–µ—Å—è—Ç–∫–∏ –∏ –µ–¥–∏–Ω–∏—Ü—ã);
–Ω–∞—Ö–æ–¥–∏—Ç—Å—è —Å—É–º–º–∞ —Ü–∏—Ñ—Ä, —Å—Ç–æ—è—â–∏—Ö –≤ –¥–≤—É—Ö —Å—Ç–∞—Ä—à–∏—Ö —Ä–∞–∑—Ä—è–¥–∞—Ö (—Å–æ—Ç–Ω–∏ –∏ –¥–µ—Å—è—Ç–∫–∏)
–≠—Ç–∏ –¥–≤–µ —Å—É–º–º—ã, –∑–∞–ø–∏—Å–∞–Ω–Ω—ã–µ –¥—Ä—É–≥ –∑–∞ –¥—Ä—É–≥–æ–º, –≤ –ø–æ—Ä—è–¥–∫–µ –Ω–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è, —Ñ–æ—Ä–º–∏—Ä—É—é—Ç –Ω–æ–≤–æ–µ —á–∏—Å–ª–æ.
–ü–æ–º–æ–≥–∏—Ç–µ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º —à–∏—Ñ—Ä–æ–≤–∞–Ω–∏—è.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–û–¥–Ω–æ —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–†–µ–∑—É–ª—å—Ç–∞—Ç —à–∏—Ñ—Ä–æ–≤–∞–Ω–∏—è'''


x = int(input())
a = x // 100
b = x // 10 % 10
c = x % 10
z = a + b
y = c + b
print(f'{max(z, y)}{min(z, y)}')
```

```python
'''–ö—Ä–∞—Å–æ—Ç–∞ —Å–ø–∞—Å—ë—Ç –º–∏—Ä
–û–¥–Ω–æ –∏–∑ –¥—Ä–µ–≤–Ω–∏—Ö –ø–æ–≤–µ—Ä–∏–π –≥–ª–∞—Å–∏—Ç, —á—Ç–æ —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ –∫—Ä–∞—Å–∏–≤–æ, –µ—Å–ª–∏ —Å—É–º–º–∞ –µ–≥–æ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π —Ü–∏—Ñ—Ä 
—Ä–∞–≤–Ω–∞ –æ—Å—Ç–∞–≤—à–µ–π—Å—è —Ü–∏—Ñ—Ä–µ —É–º–Ω–æ–∂–µ–Ω–Ω–æ–π –Ω–∞ 2.

–ù–∞–ø–∏—à–∏—Ç–µ —Å–∏—Å—Ç–µ–º—É –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é –∫—Ä–∞—Å–æ—Ç—É —á–∏—Å–ª–∞.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–û–¥–Ω–æ —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
YES ‚Äî –µ—Å–ª–∏ —á–∏—Å–ª–æ –∫—Ä–∞—Å–∏–≤–æ–µ, –∏–Ω–∞—á–µ ‚Äî NO'''


x = int(input())
a = x // 100
b = x // 10 % 10
c = x % 10
mid = (a + b + c) - (max(a, b, c) + min(a, b, c))
if (min(a, b, c) + max(a, b, c)) == mid * 2:
    print('YES')
else:
    print('NO')
```

```python
'''–ú—É–∑—ã–∫–∞–ª—å–Ω—ã–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç
–ï—Å—Ç—å –º–Ω–æ–≥–æ –º—É–∑—ã–∫–∞–ª—å–Ω—ã—Ö –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–æ–≤, –Ω–æ –í–∞—Å—è –æ–±–æ–∂–∞–µ—Ç —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫.

–£ –Ω–µ–≥–æ –∑–∞–≤–∞–ª—è–ª–æ—Å—å –Ω–µ–º–Ω–æ–≥–æ –∞–ª—é–º–∏–Ω–∏–µ–≤—ã—Ö —Ç—Ä—É–±–æ—á–µ–∫ —Ä–∞–∑–Ω–æ–π –¥–ª–∏–Ω—ã, –∏ –æ–Ω –∑–∞–¥–∞—ë—Ç—Å—è –≤–æ–ø—Ä–æ—Å–æ–º, 
–∞ –º–æ–∂–Ω–æ –ª–∏ –∏–∑ –Ω–∏—Ö —Å–¥–µ–ª–∞—Ç—å –ª—é–±–∏–º—ã–π –º—É–∑—ã–∫–∞–ª—å–Ω—ã–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–¢—Ä–∏ —á–∏—Å–ª–∞ ‚Äî –¥–ª–∏–Ω—ã —Ç—Ä—É–±–æ—á–µ–∫, –∫–∞–∂–¥–æ–µ —Å –Ω–æ–≤–æ–π —Å—Ç—Ä–æ–∫–∏.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
YES ‚Äî –µ—Å–ª–∏ –í–∞—Å–µ —É–¥–∞—Å—Ç—Å—è —Å–æ–∑–¥–∞—Ç—å –º—É–∑—ã–∫–∞–ª—å–Ω—ã–π —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫, –∏–Ω–∞—á–µ ‚Äî NO

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ
–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –≤ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–µ –ª—é–±–∞—è —Å—Ç–æ—Ä–æ–Ω–∞ –º–µ–Ω—å—à–µ —Å—É–º–º—ã –¥–≤—É—Ö –¥—Ä—É–≥–∏—Ö.'''


x = int(input())
y = int(input())
z = int(input())
mid = (x + y + z) - (min(x, y, z) + max(x, y, z))
if max(x, y, z) < mid + min(x, y, z):
    print('YES')
else:
    print('NO')
```

```python
'''–í–ª–∞—Å—Ç–µ–ª–∏–Ω –ß–∏—Å–µ–ª: –ë—Ä–∞—Ç—Å—Ç–≤–æ –æ–±—â–µ–π —Ü–∏—Ñ—Ä—ã
–í–æ –≤—Ä–µ–º–µ–Ω–∞ –º–∞–≥–∏–∏ –∏ –¥—Ä–∞–∫–æ–Ω–æ–≤ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–ª–æ –º–Ω–µ–Ω–∏–µ, —á—Ç–æ —á–∏—Å–ª–∞ –æ–±–ª–∞–¥–∞—é—Ç –≤–µ–ª–∏–∫–æ–π —Å–∏–ª–æ–π, —Å–ø–æ—Å–æ–±–Ω–æ–π –∏–∑–º–µ–Ω–∏—Ç—å –º–∏—Ä.

–í—Å—ë –Ω–∞—á–∞–ª–æ—Å—å —Å –Ω–∞–ø–∏—Å–∞–Ω–∏—è –≤–µ–ª–∏–∫–∏—Ö —á–∏—Å–µ–ª. –¢—Ä–∏ —á–∏—Å–ª–∞ –±—ã–ª–∏ –ø–µ—Ä–µ–¥–∞–Ω—ã —ç–ª—å—Ñ–∞–º. –°–µ–º—å ‚Äî –æ—Ç–¥–∞–Ω—ã –ø–æ–≤–µ–ª–∏—Ç–µ–ª—è–º –≥–Ω–æ–º–æ–≤. 
–ê –¥–µ–≤—è—Ç—å... –±—ã–ª–∏ –ø–µ—Ä–µ–¥–∞–Ω—ã —á–µ–ª–æ–≤–µ—á–µ—Å–∫–æ–º—É —Ä–æ–¥—É.

–ù–æ –≤—Å–µ –æ–Ω–∏ –æ–∫–∞–∑–∞–ª–∏—Å—å –æ–±–º–∞–Ω—É—Ç—ã, –ø–æ—Ç–æ–º—É —á—Ç–æ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–ª–æ –µ—â—ë –æ–¥–Ω–æ —á–∏—Å–ª–æ. 
–í —Å—Ç—Ä–∞–Ω–µ –ù—É–º–∏—è –Ω–∞ –±—É–º–∞–≥–µ –∏–∑ —Ç—ë–º–Ω–æ–≥–æ –ø–∞–ø–∏—Ä—É—Å–∞ –≤–ª–∞—Å—Ç–µ–ª–∏–Ω –ó–µ—Ä–æ–Ω —Ç–∞–π–Ω–æ –Ω–∞–ø–∏—Å–∞–ª –ï–¥–∏–Ω—É—é –¶–∏—Ñ—Ä—É, 
–ø–æ–¥—á–∏–Ω—è—é—â—É—é —Å–µ–±–µ –≤—Å–µ –≤–µ–ª–∏–∫–∏–µ —á–∏—Å–ª–∞.

–î–∞–≤–∞–π—Ç–µ –≤—ã—è—Å–Ω–∏–º, —á—Ç–æ —ç—Ç–æ –∑–∞ —Ü–∏—Ñ—Ä–∞.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ –¥–≤—É–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ –æ–¥–Ω–æ–≥–æ –∏–∑ —ç–ª—å—Ñ–æ–≤.
–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ ‚Äî –ì–Ω–æ–º–æ–≤.
–í —Ç—Ä–µ—Ç—å–µ–π ‚Äî –õ—é–¥–µ–π.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–û–¥–Ω–∞ —Ü–∏—Ñ—Ä–∞ ‚Äî –æ–±—â–∞—è —É –≤—Å–µ—Ö —Ç—Ä—ë—Ö —á–∏—Å–µ–ª –≤ –æ–¥–∏–Ω–∞–∫–æ–≤–æ–π –ø–æ–∑–∏—Ü–∏–∏'''


e = int(input())
g = int(input())
h = int(input())
e1 = e // 10
e2 = e % 10
g1 = g // 10
g2 = g % 10
h1 = h // 10
h2 = h % 10
if e1 == g1 == h1:
    print(e1)
elif e2 == g2 == h2:
    print(e2)
```

```python
'''–í–ª–∞—Å—Ç–µ–ª–∏–Ω –ß–∏—Å–µ–ª: –î–≤–µ –ë–∞—à–Ω–∏
–í–æ –≤—Ä–µ–º–µ–Ω–∞, –∫–æ–≥–¥–∞ –ª—é–¥–∏ –≤–µ—Ä–∏–ª–∏ –≤ –≤–µ–ª–∏–∫—É—é —Å–∏–ª—É —á–∏—Å–µ–ª, –æ–∫–∞–∑–∞–ª–æ—Å—å, —á—Ç–æ –≤–æ–ª—à–µ–±–Ω–∏–∫ –ü–∏—Ñ—É–º–∞–Ω –ø—Ä–µ–¥–∞–ª –≤—Å–µ –Ω–∞—Ä–æ–¥—ã –∏ 
—Å—Ç–∞–ª –ø–æ–º–æ–≥–∞—Ç—å –ó–µ—Ä–æ–Ω—É.

–ß—Ç–æ–±—ã –ø–æ—Å–µ—Ç–∏—Ç—å –±–∞—à–Ω–∏ –æ–±–æ–∏—Ö –∑–ª–æ–¥–µ–µ–≤ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ, –Ω–∞–º —Å–ª–µ–¥—É–µ—Ç —Ä–∞–∑–¥–µ–ª–∏—Ç—å –º–∞–≥–∏—é —á–∏—Å–ª–∞, –∫–æ—Ç–æ—Ä–æ–µ –∑–∞—â–∏—â–∞–ª–æ –Ω–∞—Å –≤ –¥–æ—Ä–æ–≥–µ.

–ß—Ç–æ–±—ã –ø–æ–¥–µ–ª–∏—Ç—å —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ, –Ω–∞–º –Ω—É–∂–Ω–æ —Å–æ—Å—Ç–∞–≤–∏—Ç—å –∏–∑ –Ω–µ–≥–æ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω—ã–µ –¥–≤—É—Ö–∑–Ω–∞—á–Ω—ã–µ —á–∏—Å–ª–∞.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–û–¥–Ω–æ —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–î–≤–∞ –∑–∞—â–∏—Ç–Ω—ã—Ö —á–∏—Å–ª–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –æ—Ç—Ä—è–¥–∞, –∑–∞–ø–∏—Å–∞–Ω–Ω—ã–µ —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª.'''


x = int(input())
a = x // 100
b = x // 10 % 10
c = x % 10
mid = (a + b + c) - (max(a, b, c) + min(a, b, c))
max_n = int(f'{max(a, b, c)}{mid}')
if min(a, b, c) != 0:
    min_n = int(f'{min(a, b, c)}{mid}')
else:
    min_n = int(f'{mid}{min(a, b, c)}')
print(min_n, max_n)
```

```python
'''–í–ª–∞—Å—Ç–µ–ª–∏–Ω –ß–∏—Å–µ–ª: –í–æ–∑–≤—Ä–∞—â–µ–Ω–∏–µ –¶–µ–∑–∞—Ä—è
–î–æ –ø–æ–±–µ–¥—ã –Ω–∞–¥ –∑–ª–æ–º –æ—Å—Ç–∞–ª—Å—è –æ–¥–∏–Ω —à–∞–≥ ‚Äî —Ä–∞–∑—Ä—É—à–∏—Ç—å –ª–æ–≥–æ–≤–æ –ó–µ—Ä–æ–Ω–∞.

–î–ª—è —ç—Ç–æ–≥–æ –Ω—É–∂–Ω–æ —Å–æ–∑–¥–∞—Ç—å —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ –º–∞–≥–∏—á–µ—Å–∫–æ–µ —á–∏—Å–ª–æ, –∫–æ—Ç–æ—Ä–æ–µ –±—É–¥–µ—Ç —Å–∏–ª—å–Ω–µ–µ –¥–≤—É—Ö –¥–≤—É—Ö–∑–Ω–∞—á–Ω—ã—Ö –∑–∞—â–∏—Ç–Ω—ã—Ö —á–∏—Å–µ–ª –ó–µ—Ä–æ–Ω–∞.

–°–∞–º—ã–π –ø—Ä–æ—Å—Ç–æ–π —Å–ø–æ—Å–æ–± —Å–æ–∑–¥–∞—Ç—å —Å–∏–ª—å–Ω–æ–µ —á–∏—Å–ª–æ:

–ø–µ—Ä–≤–æ–π –≤–∑—è—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é —Ü–∏—Ñ—Ä—É –∏–∑ –≤—Å–µ—Ö –∑–∞—â–∏—Ç–Ω—ã—Ö —á–∏—Å–µ–ª;
–ø–æ—Å–ª–µ–¥–Ω–µ–π ‚Äî –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é;
–≤ —Å–µ—Ä–µ–¥–∏–Ω—É –ø–æ—Å—Ç–∞–≤–∏—Ç—å —Å—É–º–º—É –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –±–µ–∑ —É—á—ë—Ç–∞ –ø–µ—Ä–µ–Ω–æ—Å–∞ —Ä–∞–∑—Ä—è–¥–∞.
–ü–æ–º–æ–≥–∏—Ç–µ –æ–¥–æ–ª–µ—Ç—å –∑–ª–æ.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –¥–≤—É—Ö —Å—Ç—Ä–æ–∫–∞—Ö –∑–∞–ø–∏—Å–∞–Ω—ã –∑–∞—â–∏—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –ó–µ—Ä–æ–Ω–∞.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–û–¥–Ω–æ —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ, –∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–∏–≤–µ–¥—ë—Ç –∫ –ø–æ–±–µ–¥–µ.'''


 = int(input())
g = int(input())
e1 = e // 10
e2 = e % 10
g1 = g // 10
g2 = g % 10
mid1 = (e1 + e2 + g1 + g2) - (max(e1, e2, g1, g2) + min(e1, e2, g1, g2))
mid2 = mid1 % 10
print(f'{max(e1, e2, g1, g2)}{mid2}{min(e1, e2, g1, g2)}')
```

```python
'''–õ–µ–≥–µ–Ω–¥—ã –≤–µ–ª–æ–≥–æ–Ω–æ–∫ –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç—Å—è: –∫—Ç–æ –±—ã—Å—Ç—Ä–µ–µ?
–í –Ω–æ–≤–æ–º —Å–µ–∑–æ–Ω–µ –∑–∞ –ø–µ—Ä–≤–µ–Ω—Å—Ç–≤–æ –≤ –≤–µ–ª–æ–≥–æ–Ω–∫–∞—Ö –≤–Ω–æ–≤—å –±–æ—Ä—é—Ç—Å—è –ª—É—á—à–∏–µ –∏–∑ –ª—É—á—à–∏—Ö. –ü—Ä–æ—Ç—è–∂—ë–Ω–Ω–æ—Å—Ç—å –∑–∞–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ–π —Ç—Ä–∞—Å—Å—ã ‚Äî 43872–º, 
–∏ –≤—Å–µ —Ö–æ—Ç—è—Ç –∑–Ω–∞—Ç—å, –∫—Ç–æ –ø–æ–ª—É—á–∏—Ç –∑–æ–ª–æ—Ç—É—é –º–µ–¥–∞–ª—å.

–ù–∞–º –∏–∑–≤–µ—Å—Ç–Ω—ã —Å—Ä–µ–¥–Ω–∏–µ —Å–∫–æ—Ä–æ—Å—Ç–∏ —Ç—Ä—ë—Ö –ø—Ä–µ—Ç–µ–Ω–¥–µ–Ω—Ç–æ–≤ –Ω–∞ –ø–æ–±–µ–¥—É ‚Äì –ü–µ—Ç–∏, –í–∞—Å–∏ –∏ –¢–æ–ª–∏. –ö—Ç–æ –∏–∑ –Ω–∏—Ö –ø–æ–±–µ–¥–∏—Ç?

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–∞ —Å—Ä–µ–¥–Ω—è—è —Å–∫–æ—Ä–æ—Å—Ç—å –ü–µ—Ç–∏.
–í–æ –≤—Ç–æ—Ä–æ–π ‚Äî –í–∞—Å–∏.
–í —Ç—Ä–µ—Ç—å–µ–π ‚Äî –¢–æ–ª–∏.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–ö—Ä–∞—Å–∏–≤—ã–π –ø—å–µ–¥–µ—Å—Ç–∞–ª (—à–∏—Ä–∏–Ω–∞ —Å—Ç—É–ø–µ–Ω–µ–π 8 —Å–∏–º–≤–æ–ª–æ–≤).'''


p = int(input())
v = int(input())
t = int(input())
mid = (p + v + t) - (max(p, v, t) + min(p, v, t))
if p == max(p, v, t):
    vin = '–ü–µ—Ç—è'
elif v == max(p, v, t):
    vin = '–í–∞—Å—è'
elif t == max(p, v, t):
    vin = '–¢–æ–ª—è'
if p == mid:
    sec = '–ü–µ—Ç—è'
elif v == mid:
    sec = '–í–∞—Å—è'
elif t == mid:
    sec = '–¢–æ–ª—è'
if p == min(p, v, t):
    los = '–ü–µ—Ç—è'
elif v == min(p, v, t):
    los = '–í–∞—Å—è'
elif t == min(p, v, t):
    los = '–¢–æ–ª—è'
print(f"{'': ^8}{vin: ^8}{'': ^8}")
print(f"{sec: ^8}{'': ^8}{'': ^8}")
print(f"{'': ^8}{'': ^8}{los: ^8}")
print(f'{"II": ^8}{"I": ^8}{"III": ^8}')
```

```python
'''–ö–æ—Ä–µ–Ω—å –∑–ª–∞
–ù–µ –≤—Å–µ –ª—é–±—è—Ç –º–∞—Ç–µ–º–∞—Ç–∏–∫—É, –∞ –∫—Ç–æ-—Ç–æ –¥–∞–∂–µ —Å—á–∏—Ç–∞–µ—Ç –µ—ë –Ω–∞—Å—Ç–æ—è—â–∏–º –∑–ª–æ–º –≤–æ –ø–ª–æ—Ç–∏, —Ö–æ—Ç—è –æ—Ç –Ω–µ—ë –Ω–∏–∫—É–¥–∞ –Ω–µ –¥–µ—Ç—å—Å—è. 
–ö –ø—Ä–∏–º–µ—Ä—É, Python –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ —Ä–∞–∑—Ä–∞–±–∞—Ç—ã–≤–∞–ª—Å—è —Ç–æ–ª—å–∫–æ –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö –∑–∞–¥–∞—á, 
–ø–æ—ç—Ç–æ–º—É –¥–∞–≤–∞–π—Ç–µ –∏—Å–ø–æ–ª—å–∑—É–µ–º –µ–≥–æ, —á—Ç–æ–±—ã –Ω–∞–π—Ç–∏ –∫–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í–≤–æ–¥–∏—Ç—Å—è 3 –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞ a, b, ùëê ‚Äî –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —É—Ä–∞–≤–Ω–µ–Ω–∏—è –≤–∏–¥–∞:ùëéùë•**2+ùëèùë•+ùëê=0

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–ï—Å–ª–∏ —É —É—Ä–∞–≤–Ω–µ–Ω–∏—è –Ω–µ—Ç —Ä–µ—à–µ–Ω–∏–π ‚Äî —Å–ª–µ–¥—É–µ—Ç –≤—ã–≤–µ—Å—Ç–∏ ¬´No solution¬ª.
–ï—Å–ª–∏ –∫–æ—Ä–Ω–µ–π –∫–æ–Ω–µ—á–Ω–æ–µ —á–∏—Å–ª–æ ‚Äî –∏—Ö –Ω—É–∂–Ω–æ –≤—ã–≤–µ—Å—Ç–∏ —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è —Å —Ç–æ—á–Ω–æ—Å—Ç—å—é –¥–æ —Å–æ—Ç—ã—Ö.
–ï—Å–ª–∏ –∫–æ—Ä–Ω–µ–π –Ω–µ–æ–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ ‚Äî —Å–ª–µ–¥—É–µ—Ç –≤—ã–≤–µ—Å—Ç–∏ ¬´Infinite solutions¬ª.'''


a = float(input())
b = float(input())
c = float(input())
if a == b == c == 0:
    print('Infinite solutions')
elif a == b == 0:
    print('No solution')
elif a == c == 0:
    print('0')
elif a == 0 and b != 0 and c != 0:
    x = -c / b
    print(f'{x:.2f}')
else:
    d = (b ** 2) - (4 * a * c)
    if d > 0:
        x1 = (- b + (d ** 0.5)) / (2 * a)
        x2 = (- b - (d ** 0.5)) / (2 * a)
        print(f'{min(x1, x2):.2f} {max(x1, x2):.2f}')
    elif d == 0:
        x = (- b) / 2 * a
        print(f'{x:.2f}')
    else:
        print('No solution')
```

```python
'''–¢–µ—Ä—Ä–∏—Ç–æ—Ä–∏—è –∑–ª–∞
–í –¥–∞–≤–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–∞ —Å—á–∏—Ç–∞–ª–æ—Å—å, —á—Ç–æ –µ—Å–ª–∏ –∫–∞–∫–∞—è-—Ç–æ –º–µ—Å—Ç–Ω–æ—Å—Ç—å —è–≤–ª—è–µ—Ç—Å—è —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–æ–º, —Ç–æ –≤ –Ω–µ–π –∑–∞–∫–ª—é—á–µ–Ω–æ —Å—Ç—Ä–∞—à–Ω–æ–µ –∑–ª–æ.

–ü—Ä–∏ —ç—Ç–æ–º –ª—é–¥–∏ –æ—Ü–µ–Ω–∏–≤–∞–ª–∏ —Ä–∏—Å–∫ –≤—Å—Ç—Ä–µ—Ç–∏—Ç—å –∑–ª–æ –ø–æ —Ñ–æ—Ä–º–µ —ç—Ç–æ–≥–æ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞:

–≤ –æ—Å—Ç—Ä–æ—É–≥–æ–ª—å–Ω–æ–º —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—Å—Ç—Ä–µ—Ç–∏—Ç—å –∑–ª–æ –∫—Ä–∞–π–Ω–µ –º–∞–ª–∞;
–≤ —Ç—É–ø–æ—É–≥–æ–ª—å–Ω–æ–º ‚Äî –≤–µ–ª–∏–∫–∞;
–≤ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–æ–º ‚Äî 100%.
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ –¥–ª–∏–Ω–µ —Å—Ç–æ—Ä–æ–Ω —Ç—Ä–µ—É–≥–æ–ª—å–Ω–æ–π –º–µ—Å—Ç–Ω–æ—Å—Ç–∏, –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—Å—Ç—Ä–µ—Ç–∏—Ç—å –∑–ª–æ.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–¢—Ä–∏ —á–∏—Å–ª–∞ ‚Äî –¥–ª–∏–Ω—ã —Å—Ç–æ—Ä–æ–Ω —Ç—Ä–µ—É–≥–æ–ª—å–Ω–æ–π –º–µ—Å—Ç–Ω–æ—Å—Ç–∏.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—Å—Ç—Ä–µ—Ç–∏—Ç—å –∑–ª–æ —Å–æ–≥–ª–∞—Å–Ω–æ –ø–æ–≤–µ—Ä—å—é:

–∫—Ä–∞–π–Ω–µ –º–∞–ª–∞;
–≤–µ–ª–∏–∫–∞;
100%.'''


a = int(input())
b = int(input())
c = int(input())
if (max(a, b, c) ** 2) == (min(a, b, c) ** 2 + ((a + b + c) - (min(a, b, c) + max(a, b, c))) ** 2):
    print('100%')
elif (max(a, b, c) ** 2) > (min(a, b, c) ** 2 + ((a + b + c) - (min(a, b, c) + max(a, b, c))) ** 2):
    print('–≤–µ–ª–∏–∫–∞')
else:
    print('–∫—Ä–∞–π–Ω–µ –º–∞–ª–∞')
```

```python
'''–ê–≤—Ç–æ–º–∞—Ç–∏–∑–∞—Ü–∏—è –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç–∏
–ì—Ä—É–ø–ø–∞ –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª–µ–π —Å–æ–±–∏—Ä–∞–µ—Ç—Å—è –≤—ã—Å–∞–¥–∏—Ç—å—Å—è –Ω–∞ –æ—Å—Ç—Ä–æ–≤ –Ω–µ–≤–µ—Ä–æ—è—Ç–Ω–æ —Ä–æ–≤–Ω–æ–π —Ñ–æ—Ä–º—ã, 
–Ω–æ —Ä–∞–∑–≤–µ–¥–∫–∞ –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Å–ø—É—Ç–Ω–∏–∫–∞ –≤—ã—è—Å–Ω–∏–ª–∞, —á—Ç–æ –Ω–∞ –æ—Å—Ç—Ä–æ–≤–µ –µ—Å—Ç—å –∑–æ–Ω–∞ –∑—ã–±—É—á–∏—Ö –ø–µ—Å–∫–æ–≤.

–î–ª—è –ø–æ–≤—ã—à–µ–Ω–∏—è –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç–∏ —ç–∫—Å–ø–µ–¥–∏—Ü–∏–∏ –±—ã–ª–æ —Ä–µ—à–µ–Ω–æ —Ä–∞–∑—Ä–∞–±–æ—Ç–∞—Ç—å —Å–∏—Å—Ç–µ–º—É –æ–ø–æ–≤–µ—â–µ–Ω–∏—è, 
–∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–¥—É–ø—Ä–µ–¥–∏—Ç –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª–µ–π –æ–± –æ–ø–∞—Å–Ω–æ—Å—Ç–∏. –ê –¥–ª—è —Å–Ω–∏–∂–µ–Ω–∏—è —Ä–∞—Å—Ö–æ–¥–æ–≤ –Ω–∞ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–æ –±—ã–ª–æ —Ä–µ—à–µ–Ω–æ 
–∑–∞–∫–∞–∑–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º–Ω–æ–µ –æ–±–µ—Å–ø–µ—á–µ–Ω–∏–µ.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—è, –±—É–¥–µ—Ç —Å–æ–æ–±—â–∞—Ç—å –æ –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç–∏ –≤ —ç—Ç–æ–π —Ç–æ—á–∫–µ.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–î–≤–∞ —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ ‚Äî –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—è.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–û–¥–Ω–æ –∏–∑ —Å–æ–æ–±—â–µ–Ω–∏–π:

–û–ø–∞—Å–Ω–æ—Å—Ç—å! –ü–æ–∫–∏–Ω—å—Ç–µ –∑–æ–Ω—É –∫–∞–∫ –º–æ–∂–Ω–æ —Å–∫–æ—Ä–µ–µ!
–ó–æ–Ω–∞ –±–µ–∑–æ–ø–∞—Å–Ω–∞. –ü—Ä–æ–¥–æ–ª–∂–∞–π—Ç–µ —Ä–∞–±–æ—Ç—É.
–í—ã –≤—ã—à–ª–∏ –≤ –º–æ—Ä–µ –∏ —Ä–∏—Å–∫—É–µ—Ç–µ –±—ã—Ç—å —Å—ä–µ–¥–µ–Ω–Ω—ã–º –∞–∫—É–ª–æ–π'''


x = float(input())
y = float(input())
if (x ** 2 + y ** 2) ** 0.5 > 10:
    print('–í—ã –≤—ã—à–ª–∏ –≤ –º–æ—Ä–µ –∏ —Ä–∏—Å–∫—É–µ—Ç–µ –±—ã—Ç—å —Å—ä–µ–¥–µ–Ω–Ω—ã–º –∞–∫—É–ª–æ–π!')
elif x >= 0 and y >= 0 and (x ** 2 + y ** 2) ** 0.5 <= 5:
    print('–û–ø–∞—Å–Ω–æ—Å—Ç—å! –ü–æ–∫–∏–Ω—å—Ç–µ –∑–æ–Ω—É –∫–∞–∫ –º–æ–∂–Ω–æ —Å–∫–æ—Ä–µ–µ!')
elif -4 <= x < 0 and 0 < y <= 5:
    print('–û–ø–∞—Å–Ω–æ—Å—Ç—å! –ü–æ–∫–∏–Ω—å—Ç–µ –∑–æ–Ω—É –∫–∞–∫ –º–æ–∂–Ω–æ —Å–∫–æ—Ä–µ–µ!')
elif -7 <= x < -4 and 0 < y <= 5 and (5 * x - 3 * y) > -35:
    print('–û–ø–∞—Å–Ω–æ—Å—Ç—å! –ü–æ–∫–∏–Ω—å—Ç–µ –∑–æ–Ω—É –∫–∞–∫ –º–æ–∂–Ω–æ —Å–∫–æ—Ä–µ–µ!')
elif (0.25 * x ** 2 + 0.5 * x - 9) <= y <= 0:
    print('–û–ø–∞—Å–Ω–æ—Å—Ç—å! –ü–æ–∫–∏–Ω—å—Ç–µ –∑–æ–Ω—É –∫–∞–∫ –º–æ–∂–Ω–æ —Å–∫–æ—Ä–µ–µ!')
else:
    print('–ó–æ–Ω–∞ –±–µ–∑–æ–ø–∞—Å–Ω–∞. –ü—Ä–æ–¥–æ–ª–∂–∞–π—Ç–µ —Ä–∞–±–æ—Ç—É.')
```

```python
'''–ó–∞–π–∫–∞ ‚Äî 2
–ü–æ –ø—É—Ç–∏ –¥–æ–º–æ–π —Ä–æ–¥–∏—Ç–µ–ª–∏ –≤–Ω–æ–≤—å —Ä–µ—à–∏–ª–∏ —Å—ã–≥—Ä–∞—Ç—å —Å –¥–µ—Ç—å–º–∏ –≤ –ø–æ–∏—Å–∫ –∑–≤–µ—Ä—É—à–µ–∫.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–¢—Ä–∏ —Å—Ç—Ä–æ–∫–∏ –æ–ø–∏—Å—ã–≤–∞—é—â–∏—Ö –ø—Ä–∏–¥–æ—Ä–æ–∂–Ω—É—é –º–µ—Å—Ç–Ω–æ—Å—Ç—å.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–°—Ç—Ä–æ–∫–∞ –≤ –∫–æ—Ç–æ—Ä–æ–π –µ—Å—Ç—å –∑–∞–π–∫–∞, –∞ –∑–∞—Ç–µ–º –µ—ë –¥–ª–∏–Ω–∞.
–ï—Å–ª–∏ —Ç–∞–∫–∏—Ö —Å—Ç—Ä–æ–∫ –Ω–µ—Å–∫–æ–ª—å–∫–æ, –≤—ã–±—Ä–∞—Ç—å —Ç—É, —á—Ç–æ –º–µ–Ω—å—à–µ –≤—Å–µ—Ö –ª–µ–∫—Å–∏–∫–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏.'''


text1 = input()
text2 = input()
text3 = input()
p_data = ''
if '–∑–∞–π–∫–∞' in text1:
    p_data = text1
if '–∑–∞–π–∫–∞' in text2:
    if p_data == '' or text2 < p_data:
        p_data = text2
if '–∑–∞–π–∫–∞' in text3:
    if p_data == '' or text3 < p_data:
        p_data = text3
print(p_data, len(p_data))
```
