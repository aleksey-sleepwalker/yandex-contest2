```python
'''–ê–≤—Ç–æ–º–∞—Ç–∏–∑–∞—Ü–∏—è —Å–ø–∏—Å–∫–∞
–ú–Ω–æ–≥–∏–º –≤–µ—Å—å–º–∞ —á–∞—Å—Ç–æ –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –≤–µ—Å—Ç–∏ —Å–ø–∏—Å–∫–∏ –ø—Ä–æ–¥—É–∫—Ç–æ–≤, –¥–µ–ª –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç –≤–≤–µ–¥—ë–Ω–Ω—É—é —Å—Ç—Ä–æ–∫—É –≤ –Ω—É–º–µ—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í–≤–æ–¥–∏—Ç—Å—è –æ–¥–Ω–∞ —Å—Ç—Ä–æ–∫–∞.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–¢—Ä–µ–±—É–µ—Ç—Å—è –≤—ã–≤–µ—Å—Ç–∏ –Ω—É–º–µ—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫, —Å–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–π –∏–∑ –µ—ë —Å–ª–æ–≤.'''


text = input().split()
for index, value in enumerate(text, 1):
    print(f'{index}. {value}')
```

```python
'''–°–±–æ—Ä—ã –Ω–∞ –ø—Ä–æ–≥—É–ª–∫—É
–í–æ—Å–ø–∏—Ç–∞—Ç–µ–ª—å –≤ –¥–µ—Ç—Å–∫–æ–º —Å–∞–¥—É —É—Å—Ç–∞–ª —Ç—Ä–∞—Ç–∏—Ç—å –≤—Ä–µ–º—è, —á—Ç–æ–±—ã –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–µ—Ç–µ–π –ø–æ –ø–∞—Ä–∞–º.

–û–Ω –¥–æ–≥–æ–≤–æ—Ä–∏–ª—Å—è —Å –¥–µ—Ç—å–º–∏, —á—Ç–æ–±—ã —Ç–µ –¥–µ–ª–∏–ª–∏—Å—å –Ω–∞ –¥–≤–µ, –ø–æ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ —Ä–∞–≤–Ω—ã–µ, –≥—Ä—É–ø–ø—ã.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ —Å–ø–∏—Å–∫—É –¥–≤—É—Ö —à–µ—Ä–µ–Ω–≥ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –ø–∞—Ä—ã –¥–µ—Ç–µ–π.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í–≤–æ–¥–∏—Ç—Å—è –¥–≤–µ —Å—Ç—Ä–æ–∫–∏ —Å –∏–º–µ–Ω–∞–º–∏ –¥–µ—Ç–µ–π, –∑–∞–ø–∏—Å–∞–Ω–Ω—ã–º–∏ —á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é –∏ –ø—Ä–æ–±–µ–ª.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–¢—Ä–µ–±—É–µ—Ç—Å—è –≤—ã–≤–µ—Å—Ç–∏ —Å–ø–∏—Å–æ–∫ –ø–∞—Ä, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–Ω–æ —Å–æ—Å—Ç–∞–≤–∏—Ç—å, –µ—Å–ª–∏ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –±—Ä–∞—Ç—å –∏–∑ –∫–∞–∂–¥–æ–π —à–µ—Ä–µ–Ω–≥–∏ –ø–æ –æ–¥–Ω–æ–º—É —Ä–µ–±—ë–Ω–∫—É.
–ò–º–µ–Ω–∞ –≤ –ø–∞—Ä–∞—Ö –≤—ã–≤–æ–¥–∏—Ç—å —á–µ—Ä–µ–∑ –¥–µ—Ñ–∏—Å –æ–∫—Ä—É–∂—ë–Ω–Ω—ã–π –ø—Ä–æ–±–µ–ª–∞–º–∏.'''


groupe1 = input().split(', ')
groupe2 = input().split(', ')
pairs = list(zip(groupe1, groupe2))
for i in pairs:
    print(f'{i[0]} - {i[1]}')
```

```python
'''–†–∞—Ü–∏–æ–Ω–∞–ª—å–Ω–∞—è —Å—á–∏—Ç–∞–ª–æ—á–∫–∞
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç —Å—á—ë—Ç –ø–æ –∑–∞–¥–∞–Ω–Ω—ã–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª –≤–≤–æ–¥—è—Ç—Å—è 3 —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ ‚Äî –Ω–∞—á–∞–ª–æ —Å—á–µ—Ç–∞, –∫–æ–Ω–µ—Ü –∏ —à–∞–≥.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏.'''


from itertools import count
data = input().split()
start = float(data[0])
end = float(data[1])
step = float(data[2])
for i in count(start, step):
    if i <= end:
        print(f'{i:.2f}')
    else:
        break
```

```python
'''–°–ª–æ–≤–∞—Ä–Ω–∞—è —ë–ª–∫–∞
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Å—Ç—Ä–æ–∫—É —Å–ª–æ–≤ –≤ —ë–ª–∫—É –∫–∞–∫ –ø–æ–∫–∞–∑–∞–Ω–æ –≤ –ø—Ä–∏–º–µ—Ä–µ.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª –≤–≤–æ–¥—è—Ç—Å—è —Å–ª–æ–≤–∞ —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã–µ –ø—Ä–æ–±–µ–ª–æ–º.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–ù–µ—Å–∫–æ–ª—å–∫–æ —Å—Ç—Ä–æ–∫. –í –∫–∞–∂–¥–æ–π —Å–ª–µ–¥—É—é—â–µ–π —Å—Ç—Ä–æ–∫–µ –Ω–∞ –æ–¥–Ω–æ —Å–ª–æ–≤–æ –±–æ–ª—å—à–µ.'''


from itertools import accumulate
text = [word + ' ' for word in input().split()]
for i in accumulate(text):
    print(i)
```

```python
'''–°–ø–∏—Å–æ–∫ –ø–æ–∫—É–ø–æ–∫
–ü–æ—Ö–æ–¥ –≤ –º–∞–≥–∞–∑–∏–Ω —á–∞—Å—Ç–æ –≤—ã–∑—ã–≤–∞–µ—Ç –ø—Ä–æ–±–ª–µ–º—ã. –ï—Å–ª–∏ –Ω–µ –ø–æ–¥–≥–æ—Ç–æ–≤–∏—Ç—å —Å–ø–∏—Å–æ–∫, –º–æ–∂–Ω–æ —É–π—Ç–∏ –≤ –º–∞–≥–∞–∑–∏–Ω –∑–∞ —Ö–ª–µ–±–æ–º, 
–∞ –≤–µ—Ä–Ω—É—Ç—å—Å—è —Å –¥–µ—Å—è—Ç–∫–æ–º –ø–∞–∫–µ—Ç–æ–≤. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–±–∏—Ä–∞–µ—Ç –ø–æ–∂–µ–ª–∞–Ω–∏—è —Å–µ–º—å–∏ (–º–∞–º—ã, –ø–∞–ø—ã –∏ –¥–æ—á–∫–∏) 
–≤ –µ–¥–∏–Ω—ã–π —Å–ø–∏—Å–æ–∫.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í —Ç—Ä—ë—Ö —Å—Ç—Ä–æ–∫–∞—Ö –∑–∞–ø–∏—Å–∞–Ω—ã –∂–µ–ª–∞–µ–º—ã–µ –ø—Ä–æ–¥—É–∫—Ç—ã (—á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é –∏ –ø—Ä–æ–±–µ–ª).

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É —Å–ø–∏—Å–æ–∫ –ø—Ä–æ–¥—É–∫—Ç–æ–≤ —Å –Ω—É–º–µ—Ä–∞—Ü–∏–µ–π.'''


from itertools import chain
products = sorted(set(chain.from_iterable([input().split(', ') for _ in range(3)])))
for index, value in enumerate(products, 1):
    print(f'{index}. {value}')
```

```python
'''–ö–æ–ª–æ–¥–∞ –∫–∞—Ä—Ç
–°–≥–µ–Ω–µ—Ä–∏—Ä—É–π—Ç–µ –∏ –≤—ã–≤–µ–¥–∏—Ç–µ –Ω–∞ —ç–∫—Ä–∞–Ω –∫–æ–ª–æ–¥—É –∏–≥—Ä–∞–ª—å–Ω—ã—Ö –∫–∞—Ä—Ç –±–µ–∑ –æ–¥–Ω–æ–π –∏–∑ –º–∞—Å—Ç–µ–π: —Ç—Ä–∏ –º–∞—Å—Ç–∏ –æ—Ç –¥–≤–æ–π–∫–∏ –¥–æ —Ç—É–∑–∞. 
–ú–∞—Å—Ç—å, –∫–æ—Ç–æ—Ä–∞—è –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –≤—ã–±—Ä–æ—à–µ–Ω–∞, –≤–≤–æ–¥–∏—Ç—Å—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º. 
–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∏—Ç–µ—Ä–∞—Ç–æ—Ä product –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã —Å–∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞—Ç—å –º–∞—Å—Ç–∏ –∏ –Ω–æ–º–∏–Ω–∞–ª—ã.

–ù–∞–∑–≤–∞–Ω–∏—è –∫–∞—Ä—Ç –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –≤ —Ñ–æ—Ä–º–∞—Ç–µ ¬´–Ω–æ–º–∏–Ω–∞–ª –º–∞—Å—Ç—å¬ª. –ù–∞–ø—Ä–∏–º–µ—Ä: ¬´2 –ø–∏–∫¬ª, ¬´10 —Ç—Ä–µ—Ñ¬ª, ¬´–≤–∞–ª–µ—Ç –±—É–±–µ–Ω¬ª, ¬´–¥–∞–º–∞ —á–µ—Ä–≤–µ–π¬ª.
–ü–æ—Ä—è–¥–æ–∫ –º–∞—Å—Ç–µ–π –∏–º–µ–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–ù–∞–∑–≤–∞–Ω–∏–µ –º–∞—Å—Ç–∏, –∫–æ—Ç–æ—Ä–∞—è –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –≤—ã–±—Ä–æ—à–µ–Ω–∞, –ø–µ—Ä–µ–¥–∞—ë—Ç—Å—è –≤ —Ç–æ–º –∂–µ –ø–∞–¥–µ–∂–µ, –≤ –∫–æ—Ç–æ—Ä–æ–º –æ–Ω–∞ –≤—ã–≤–æ–¥–∏—Ç—Å—è –Ω–∞ —ç–∫—Ä–∞–Ω: 
¬´–ø–∏–∫¬ª, ¬´—Ç—Ä–µ—Ñ¬ª, ¬´–±—É–±–µ–Ω¬ª –∏ ¬´—á–µ—Ä–≤–µ–π¬ª.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–°–ø–∏—Å–æ–∫ –∫–∞—Ä—Ç –≤ –∫–æ–ª–æ–¥–µ –ø–æ —É–≤–µ–ª–∏—á–µ–Ω–∏—é –Ω–æ–º–∏–Ω–∞–ª–∞, –∑–∞—Ç–µ–º –º–∞—Å—Ç–∏ (–∫–∞–∫ –≤ –ø—Ä–µ—Ñ–µ—Ä–∞–Ω—Å–µ).'''


from itertools import product
nom = '2 3 4 5 6 7 8 9 10 –≤–∞–ª–µ—Ç –¥–∞–º–∞ –∫–æ—Ä–æ–ª—å —Ç—É–∑'.split()
mas = '–ø–∏–∫ —Ç—Ä–µ—Ñ –±—É–±–µ–Ω —á–µ—Ä–≤–µ–π'.split()
kol = list(product(nom, mas))
ask = input()
for i in kol:
    if i[1] == ask:
        continue
    else:
        print(i[0], i[1])
```

```python
'''–ò–≥—Ä–æ–≤–∞—è —Å–µ—Ç–∫–∞
–†–µ–±—è—Ç–∞ –≤ –∫–ª–∞—Å—Å–µ —Ä–µ—à–∏–ª–∏ —É—Å—Ç—Ä–æ–∏—Ç—å —á–µ–º–ø–∏–æ–Ω–∞—Ç –ø–æ —à–∞—à–∫–∞–º. –ö —Å–æ–∂–∞–ª–µ–Ω–∏—é, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—á–µ–Ω–∏–∫–æ–≤ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —Å—Ç–µ–ø–µ–Ω—å—é –¥–≤–æ–π–∫–∏
–∏ –ø–æ—ç—Ç–æ–º—É –∫–ª–∞—Å—Å–∏—á–µ—Å–∫—É—é —Ç—É—Ä–Ω–∏—Ä–Ω—É—é —Å–µ—Ç–∫—É —Å–æ–∑–¥–∞—Ç—å —Å–ª–æ–∂–Ω–æ. 
–î–ª—è –≤—ã—è–≤–ª–µ–Ω–∏—è —Ñ–∞–≤–æ—Ä–∏—Ç–æ–≤ —Ä–µ–±—è—Ç–∞ –¥–æ–≥–æ–≤–æ—Ä–∏–ª–∏—Å—å —Å—ã–≥—Ä–∞—Ç—å –ø–æ –ø—Ä–∏–Ω—Ü–∏–ø—É ¬´–∫–∞–∂–¥—ã–π —Å –∫–∞–∂–¥—ã–º¬ª. 
–ü—Ä–æ–¥—É–º–∞–π—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç —Å–ø–∏—Å–æ–∫ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –∏–≥—Ä.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ —á–∏—Å–ª–æ —É—á–µ–Ω–∏–∫–æ–≤ (N).
–í –∫–∞–∂–¥–æ–π –∏–∑ –ø–æ—Å–ª–µ–¥—É—é—â–∏—Ö N —Å—Ç—Ä–æ–∫ –∑–∞–ø–∏—Å–∞–Ω–æ –æ–¥–Ω–æ –∏–º—è.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–°–ø–∏—Å–æ–∫ –∏–≥—Ä –≤ —Ñ–æ—Ä–º–∞—Ç–µ:
<–ò–≥—Ä–æ–∫ 1> - <–ò–≥—Ä–æ–∫ 2>
–ü–æ—Ä—è–¥–æ–∫ –∏–≥—Ä –Ω–µ –∏–º–µ–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è.'''


from itertools import combinations
players = []
for _ in range(int(input())):
    players.append(input())
for i in list(combinations(players, 2)):
    print(f'{i[0]} - {i[1]}')
```

```python
'''–ú–µ–Ω—é –ø–∏—Ç–∞–Ω–∏—è 2.0
–í –¥–µ—Ç—Å–∫–æ–º —Å–∞–¥—É –µ–∂–µ–¥–Ω–µ–≤–Ω–æ –ø–æ–¥–∞—é—Ç –Ω–æ–≤—É—é –∫–∞—à—É –Ω–∞ –∑–∞–≤—Ç—Ä–∞–∫.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∏—Ç —Ä–∞—Å–ø–∏—Å–∞–Ω–∏–µ –∫–∞—à –Ω–∞ –±–ª–∏–∂–∞–π—à–∏–µ –¥–Ω–∏.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í–≤–æ–¥–∏—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ M ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–∞—à –≤ –º–µ–Ω—é. 
–í –∫–∞–∂–¥–æ–π –∏–∑ –ø–æ—Å–ª–µ–¥—É—é—â–∏—Ö M —Å—Ç—Ä–æ–∫ –∑–∞–ø–∏—Å–∞–Ω–æ –æ–¥–Ω–æ –Ω–∞–∑–≤–∞–Ω–∏–µ –∫–∞—à–∏. –í –∫–æ–Ω—Ü–µ –ø–µ—Ä–µ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ N ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–í—ã–≤–µ—Å—Ç–∏ —Å–ø–∏—Å–æ–∫ –∫–∞—à –≤ –ø–æ—Ä—è–¥–∫–µ –ø–æ–¥–∞—á–∏.'''


from itertools import cycle
menu = [input() for _ in range(int(input()))]
days = int(input())
s = 0
for i in cycle(menu):
    if s < days:
        print(i)
    else:
        break
    s += 1
```

```python
'''–¢–∞–±–ª–∏—Ü–∞ —É–º–Ω–æ–∂–µ–Ω–∏—è 3.0
–ú–µ—Å—Ç–Ω–∞—è —Ñ–∞–±—Ä–∏–∫–∞ –∫–∞–Ω—Ü–µ–ª—è—Ä—Å–∫–∏—Ö —Ç–æ–≤–∞—Ä–æ–≤ –∑–∞–∫–∞–∑–∞–ª–∞ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç —Ç–∞–±–ª–∏—Ü—ã —É–º–Ω–æ–∂–µ–Ω–∏—è.
–î–∞–≤–∞–π—Ç–µ –ø–æ–º–æ–∂–µ–º –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—é.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í–≤–æ–¥–∏—Ç—Å—è –æ–¥–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ ‚Äî —Ç—Ä–µ–±—É–µ–º—ã–π —Ä–∞–∑–º–µ—Ä —Ç–∞–±–ª–∏—Ü—ã.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–¢–∞–±–ª–∏—Ü–∞ —É–º–Ω–æ–∂–µ–Ω–∏—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞.'''


from itertools import product, islice

size = int(input())
n = [i for i in range(1, size + 1)]
table = [x * y for x, y in product(n, repeat=2)]
for j in range(size):
    print(*islice(table, j * size, (j + 1) * size))
```

```python
'''–ú—ã –¥–µ–ª–∏–ª–∏ –∞–ø–µ–ª—å—Å–∏–Ω 2.0
–ê–Ω—è, –ë–æ—Ä—è –∏ –í–æ–≤–∞ —Ä–µ—à–∏–ª–∏ —Å—ä–µ—Å—Ç—å –∞–ø–µ–ª—å—Å–∏–Ω.
–ü–æ–¥—Å–∫–∞–∂–∏—Ç–µ —Ä–µ–±—è—Ç–∞–º, –∫–∞–∫ –∏–º –µ–≥–æ —Ä–∞–∑–¥–µ–ª–∏—Ç—å.

–†–∞–∑—Ä–∞–±–æ—Ç–∞–π—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –≤—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ —Å–ø–æ—Å–æ–±—ã —Ä–∞–∑–¥–µ–ª–∫–∏ –∞–ø–µ–ª—å—Å–∏–Ω–∞.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–æ–ª–µ–∫ –∞–ø–µ–ª—å—Å–∏–Ω–∞ (
ùëÅ
N).

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–¢–∞–±–ª–∏—Ü–∞ –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –∞–ø–µ–ª—å—Å–∏–Ω–∞.

–ü—Ä–∏–º–µ—á–∞–Ω–∏—è
–ö–∞–∂–¥–æ–º—É —Ä–µ–±—ë–Ω–∫—É –¥–æ–ª–∂–Ω–∞ –¥–æ—Å—Ç–∞—Ç—å—Å—è —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∞ –¥–æ–ª—å–∫–∞ –∞–ø–µ–ª—å—Å–∏–Ω–∞.
–ù–∏ –æ–¥–Ω–æ–π –¥–æ–ª—å–∫–∏ –Ω–µ –¥–æ–ª–∂–Ω–æ –æ—Å—Ç–∞—Ç—å—Å—è.
–í—ã–≤–æ–¥–∏—Ç—å –≤–∞—Ä–∏–∞–Ω—Ç—ã –≤ –ø–æ—Ä—è–¥–∫–µ —É–≤–µ–ª–∏—á–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥–æ–ª–µ–∫ —É –ê–Ω–∏, —Å–ª–µ–¥–æ–º –ë–æ—Ä–∏ –∏ –∑–∞—Ç–µ–º –í–æ–≤—ã.'''


from itertools import combinations

pices = int(input())
s_pices = [i for i in range(1, pices)]
print('–ê –ë –í')
for x, y in list(combinations(s_pices, 2)):
    print(x, y - x, pices - y)
```

```python
'''–ß–∏—Å–ª–æ–≤–æ–π –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ 3.0
–†–µ–±—è—Ç–∞ –≤ –¥–µ—Ç—Å–∫–æ–º —Å–∞–¥—É –≤–Ω–æ–≤—å —É—á–∞—Ç—Å—è —Å—á–∏—Ç–∞—Ç—å, –∏ –≤–æ—Å–ø–∏—Ç–∞—Ç–µ–ª—å–Ω–∏—Ü–∞ —Ä–µ—à–∏–ª–∞ —Å–¥–µ–ª–∞—Ç—å —Ç–∞–∫, 
—á—Ç–æ–±—ã –∏–º –±—ã–ª–æ –ø—Ä–æ—â–µ –æ—Å–≤–æ–∏—Ç—å –Ω–æ–≤—ã–π –Ω–∞–≤—ã–∫. –î–ª—è —ç—Ç–æ–≥–æ –æ–Ω–∞ —Ö–æ—á–µ—Ç –æ—Ñ–æ—Ä–º–∏—Ç—å —Å–ø–∏—Å–æ–∫ –∏–∑—É—á–∞–µ–º—ã—Ö —á–∏—Å–µ–ª –æ—Å–æ–±—ã–º –æ–±—Ä–∞–∑–æ–º.
–î–µ—Ç–∏ —Å–ø—Ä–∞–≤–ª—è—é—Ç—Å—è –≤–µ—Å—å–º–∞ –±—ã—Å—Ç—Ä–æ, –ø–æ—ç—Ç–æ–º—É –µ–π —Ç—Ä–µ–±—É–µ—Ç—Å—è –ø—Ä–æ–≥—Ä–∞–º–º–∞, –∫–æ—Ç–æ—Ä–∞—è —Å–ø–æ—Å–æ–±–Ω–∞ —Å—Ç—Ä–æ–∏—Ç—å —á–∏—Å–ª–æ–≤—ã–µ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∏. 
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∏—Ç —á–∏—Å–ª–æ–≤–æ–π –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ —Ç—Ä–µ–±—É–µ–º–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ —á–∏—Å–ª–æ N ‚Äî –≤—ã—Å–æ—Ç–∞ —á–∏—Å–ª–æ–≤–æ–≥–æ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞.
–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ —É–∫–∞–∑–∞–Ω–æ —á–∏—Å–ª–æ M ‚Äî —à–∏—Ä–∏–Ω–∞ —á–∏—Å–ª–æ–≤–æ–≥–æ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–ù—É–∂–Ω–æ –≤—ã–≤–µ—Å—Ç–∏ —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —á–∏—Å–ª–æ–≤–æ–π –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ —Ç—Ä–µ–±—É–µ–º–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞.
–ß—Ç–æ–±—ã –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –±—ã–ª –∫—Ä–∞—Å–∏–≤—ã–º, –∫–∞–∂–¥—ã–π –µ–≥–æ —Å—Ç–æ–ª–±–µ—Ü –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤–æ–π —à–∏—Ä–∏–Ω—ã.'''


from itertools import product, islice

n = int(input())
m = int(input())
cell = len(str(n * m))
for i, j in product(range(1, n + 1), range(1, m + 1)):
    print(f'{((i - 1) * m + j):>{cell}}', end=' ')
    if j == m:
        print()
```

```python
'''–°–ø–∏—Å–æ–∫ –ø–æ–∫—É–ø–æ–∫ 2.0
–î–∞–≤–∞–π—Ç–µ –≤–Ω–æ–≤—å –ø–æ–º–æ–∂–µ–º —á–µ–ª–æ–≤–µ–∫—É —Å –ø–æ–∫—É–ø–∫–∞–º–∏. –†–∞–∑—Ä–∞–±–æ—Ç–∞–π—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–±–∏—Ä–∞–µ—Ç –ø–æ–∂–µ–ª–∞–Ω–∏—è —Å–µ–º—å–∏ –≤ –µ–¥–∏–Ω—ã–π —Å–ø–∏—Å–æ–∫.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–¥–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ N ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–ª–µ–Ω–æ–≤ —Å–µ–º—å–∏. 
–í —Å–ª–µ–¥—É—é—â–∏—Ö N —Å—Ç—Ä–æ–∫–∞—Ö –∑–∞–ø–∏—Å–∞–Ω—ã –∂–µ–ª–∞–µ–º—ã–µ –ø—Ä–æ–¥—É–∫—Ç—ã (—á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é –∏ –ø—Ä–æ–±–µ–ª).

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É —Å–ø–∏—Å–æ–∫ –ø—Ä–æ–¥—É–∫—Ç–æ–≤ —Å –Ω—É–º–µ—Ä–∞—Ü–∏–µ–π.'''


products = []
for _ in range(int(input())):
    products += input().split(', ')
for index, value in enumerate(sorted(products), start=1):
    print(f'{index}. {value}')
```

```python
'''–†–∞—Å—Å—Ç–∞–Ω–æ–≤–∫–∞ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤
–†–∞—Å—Å—Ç–∞–Ω–æ–≤–∫–∞ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤ –Ω–∞ —Å—Ç–∞—Ä—Ç–µ ‚Äî —Å–ª–æ–∂–Ω–∞—è –∑–∞–¥–∞—á–∞. –û–¥–Ω–∞–∫–æ –ø—Ä–∏ –ø–æ–º–æ—â–∏ –∏—Ç–µ—Ä–∞—Ç–æ—Ä–æ–≤ –æ–Ω–∞ —Ä–µ—à–∞–µ—Ç—Å—è –∑–∞ –ø–∞—Ä—É —Å—Ç—Ä–æ–∫. 
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç —Å–ø–∏—Å–æ–∫ –≤–æ–∑–º–æ–∂–Ω—ã—Ö —Ä–∞—Å—Å—Ç–∞–Ω–æ–≤–æ–∫ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤ –Ω–∞ —Å—Ç–∞—Ä—Ç–µ.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–¥–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ N ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤. 
–í —Å–ª–µ–¥—É—é—â–∏—Ö N —Å—Ç—Ä–æ–∫–∞—Ö –∑–∞–ø–∏—Å–∞–Ω—ã –∏–º–µ–Ω–∞ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É —Å–ø–∏—Å–æ–∫ —Ä–∞—Å—Å—Ç–∞–Ω–æ–≤–æ–∫.
–ò–º–µ–Ω–∞ –≤ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ –≤—ã–≤–æ–¥–∏—Ç—å —á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é –∏ –ø—Ä–æ–±–µ–ª.'''


from itertools import permutations

players = [input() for _ in range(int(input()))]
for i in sorted(list(permutations(players))):
    print(', '.join(i))
```

```python
'''–°–ø–æ—Ä—Ç–∏–≤–Ω—ã–µ –≥–∞–¥–∞–Ω–∏—è
–•–æ—Ä–æ—à–æ, —Å–ø–æ—Ä—Ç—Å–º–µ–Ω—ã —Ä–∞—Å—Å—Ç–∞–≤–ª–µ–Ω—ã –Ω–∞ —Å—Ç–∞—Ä—Ç–µ. –í–æ—Ç —Ç–æ–ª—å–∫–æ —É–≥–∞–¥–∞—Ç—å —Ñ–∏–Ω–∞–ª–∏—Å—Ç–æ–≤ –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ. 
–î–∞–≤–∞–π—Ç–µ –Ω–∞–ø–∏—à–µ–º –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç —Å–ø–∏—Å–æ–∫ –≤–æ–∑–º–æ–∂–Ω—ã—Ö –ø–æ–±–µ–¥–∏—Ç–µ–ª–µ–π.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–¥–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ N ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤. 
–í —Å–ª–µ–¥—É—é—â–∏—Ö N —Å—Ç—Ä–æ–∫–∞—Ö –∑–∞–ø–∏—Å–∞–Ω—ã –∏–º–µ–Ω–∞ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–æ–≤.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É —Å–ø–∏—Å–æ–∫ –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤.
–ò–º–µ–Ω–∞ –≤ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ –≤—ã–≤–æ–¥–∏—Ç—å —á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é –∏ –ø—Ä–æ–±–µ–ª.'''


from itertools import permutations

players = [input() for _ in range(int(input()))]
for i in sorted(list(permutations(players, 3))):
    print(', '.join(i))
```

```python
'''–°–ø–∏—Å–æ–∫ –ø–æ–∫—É–ø–æ–∫ 3.0
–í —ç—Ç–æ—Ç —Ä–∞–∑ —Å–µ–º—å—è –¥–æ–≥–æ–≤–æ—Ä–∏–ª–∞—Å—å –æ —Ç–æ–º, —á—Ç–æ –≤ —Ü–µ–ª—è—Ö —ç–∫–æ–Ω–æ–º–∏–∏ –±—é–¥–∂–µ—Ç–∞, –æ–Ω–∏ –±—É–¥—É—Ç —Å–æ–≤–µ—Ä—à–∞—Ç—å –≤ –¥–µ–Ω—å —Ç–æ–ª—å–∫–æ —Ç—Ä–∏ –ø–æ–∫—É–ø–∫–∏. 
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≥–æ—Ç–æ–≤–∏—Ç –≤–∞—Ä–∏–∞–Ω—Ç—ã —Å–ø–∏—Å–∫–æ–≤ –ø–æ–∫—É–ø–æ–∫.

–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–¥–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ N ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–ª–µ–Ω–æ–≤ —Å–µ–º—å–∏. 
–í —Å–ª–µ–¥—É—é—â–∏—Ö N —Å—Ç—Ä–æ–∫–∞—Ö –∑–∞–ø–∏—Å–∞–Ω—ã –∂–µ–ª–∞–µ–º—ã–µ –ø—Ä–æ–¥—É–∫—Ç—ã (—á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é –∏ –ø—Ä–æ–±–µ–ª).

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞
–í–∞—Ä–∏–∞–Ω—Ç—ã —Å–ø–∏—Å–∫–æ–≤ –ø–æ–∫—É–ø–æ–∫ –≤ –∞–ª—Ñ–∞–≤–∏—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ.'''


from itertools import permutations

products = []
for _ in range(int(input())):
    products += input().split(', ')
for i in sorted(list(permutations(products, 3))):
    print(' '.join(i))
```
